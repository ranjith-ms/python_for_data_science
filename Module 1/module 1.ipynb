#Global Variables
Variables that are created outside of a function are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.

```
#example
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

If you create a variable with the same name inside a function, this variable will be local, and can only be used inside the function. The global variable with the same name will remain as it was, global and with the original value.

```
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

#The global Keyword
Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.

To create a global variable inside a function, you can use the **global** keyword.

```
#example
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

#Arithmetic operators and expressions

Operators are special symbols that are useful for doing computations like addition, subtraction, multiplication, division, and exponentiation etc. The operators are always applied to some values which are called operands.

1)+ =>Addition	=>	a + b

2)-   =>Subtraction =>a – b 

3)* => Multiplication	=>	a * b 

4)/ =>Division	=>	b / a 

5)% => Modulus 	b % a

6)** => Exponent(power) =>	a**b 

7)//	=>Floor Division =>a//b

```
#assigning values to variables
a = 21
b = 10
c = 0
```

**Addition** :	Adds values on either side of the operator.

```
#addition example
#here + is the operation a and b variables are operands
c = a + b
print(c)
```

**Subtraction:**	Subtracts right hand operand from left hand operand.

```
#subtraction example
#here - is the operation a and b variables are operands
c = a - b
print(c) 
```

**Multiplication:**	Multiplies values on either side of the operator

```
#multiplication example
#here * is the operation a and b variables are operands
c = a * b
print(c)
```

**Division:**	Divides left hand operand by right hand operand

```
#division example
#here / is the operation a and b variables are operands
c = a / b
print(c)
```

**Floor division:** is a normal division operation except that it returns the largest possible integer. 

```
#floor division example
#here // is the operation a and b variables are operands
c = a // b
print(c)
```

**Modulus:**	Divides left hand operand by right hand operand and returns remainder

```
#modulus example
#here % is the operation a and b variables are operands
c = a % b
print(c)
```

**Exponent:**	Performs exponential (power) calculation on operators

```
#exponential example
#here ** is the operation a and b variables are operands
a = 2
b = 3
c = a**b 
print(c)
```
